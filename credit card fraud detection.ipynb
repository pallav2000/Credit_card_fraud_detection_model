# Project Description :- Credit Card Fraud Detection

# About the Dataset
This is a simulated credit card transaction dataset containing legitimate and fraud transactions from the duration 1st Jan 2019 - 31st Dec 2020. It covers credit cards of 1000 customers doing transactions with a pool of 800 merchants.

# About the Columns
This dataset contains total 22 columns which are:-
1. trans_date_trans_time - this column shows the exact date and column on which transaction is being done
2. cc_num - this column contains different credit cards number generated by bank.
3. merchant - this column contains 800 merchants who buys a credit cards in the bank.
4. category -  there are 14 unique category in which credit card is used.
5. amt - this column shows all the money spent on the respective jobs
6. first - this column contains first name of person doing the transaction
7. last - this column contains last name of person doing the transaction
8. gender - this column contains gender of person doing the transaction
9. street - this column shows the street address where the transaction is being carried out
10. city - this column shows the city address where the transaction is being carried out
11. state- this column shows the state address where the transaction is being carried out
12. zip- this column shows the zip code of  where the transaction is being carried out
13. lat- this column shows the  latitude where the transaction is being carried out
14. long - this column shows the longitude where the transaction is being carried out
15. city_pop - this column shows the population of the city where transaction is being made
16. job- this  column shows the motive behind the transaction like for consulting,recharge,etc.
17. dob - this column shows the date of birth of the person who is doing the transaction 
18. trans_num- this column show the unique transaction number generated by bank for every transaction.
19. unix_time - this column shows the total time elapsed after a particular transaction
20. merch_lat - this column shows the latitude of the merchant
21. merch_long - this column shows the longitude of the merchant
22. is_fraud - this column shows whether a particular transaction is fraud or not
